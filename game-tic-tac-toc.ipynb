```python
def create_game_board():
    # Create a 3x3 game board with empty squares
    board = [['', '', ''], ['', '', ''], ['', '', '']]
    return board
```

```python
def print_game_board(board):
    print("-------------")
    square_number = 1
    for row in board:
        print("|", end="")
        for square in row:
            symbol = square if square else square_number
            print(f" {symbol} |", end="")
            square_number += 1
        print("\n-------------")
```

```python
def handle_player_move(board, player_symbol):
    while True:
        try:
            square_number = int(input(f"Enter the square number (1-9) for Player {player_symbol}: "))

            if 1 <= square_number <= 9:
                row = (square_number - 1) // 3
                col = (square_number - 1) % 3

                if board[row][col] == '':
                    board[row][col] = player_symbol
                    break
                else:
                    print("That square is already occupied. Please choose another square.")
            else:
                print("Please enter a valid square number (1-9).")
        except ValueError:
            print("Invalid input. Please enter a valid integer for the square number.")

```
